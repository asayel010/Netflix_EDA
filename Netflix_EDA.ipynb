```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
df = pd.read_csv("netflix_titles.csv")
df.head()
```

The dataset contains more Movies than TV Shows, indicating that Netflix’s catalog is primarily movie-focused.

The United States appears to have the highest number of titles, followed by other major content-producing countries.

There is a clear increase in content production in recent years, showing Netflix’s expansion over time.

Most content falls under TV-MA and TV-14 ratings, suggesting the platform targets mature audiences.

```python
df.shape
df.columns
df.info()
df.isnull().sum()
df.describe()
df.describe(include='object')
sns.countplot(data=df, x='type')
plt.title("Movies vs TV Shows")
plt.show()
df['country'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Countries")
plt.show()
df['release_year'].value_counts().sort_index().plot(kind='line')
plt.title("Content Released Per Year")
plt.show()
sns.countplot(data=df, x='rating')
plt.xticks(rotation=45)
plt.title("Content Rating Distribution")
plt.show()
```

This exploratory data analysis provided insights into the structure and distribution of Netflix content. 
The dataset is dominated by movies, with the majority of content coming from the United States. 
Content production has significantly increased in recent years, and most titles are rated for mature audiences.
